```python
# Import Libraries 
from plant_sqlite import select_plants, select_plants_data
from plantcv import plantcv as pcv
import cv2
import json
import numpy as np
from matplotlib import pyplot as plt
```

```python
plants = select_plants()
plants_data = select_plants_data()
```

```python
for images in plants:
	process = np.concatenate(images, axis=1)
	plt.figure(figsize=(60, 660))
	plt.imshow(cv2.cvtColor(process, cv2.COLOR_BGR2RGB))
	plt.show()
```

```python
print(plants_data[0])
```

```python
for index in range(len(plants_data)):
	rdata = json.loads(plants_data[index])
	branch_pts = len(rdata["default"]["branch_pts"]["value"])
	tips = len(rdata["default"]["tips"]["value"])
	area = rdata["default"]["area"]["value"]
	solidity = rdata["default"]["solidity"]["value"]
	perimeter = rdata["default"]["perimeter"]["value"]
	width = rdata["default"]["width"]["value"]
	height = rdata["default"]["height"]["value"]
	longest_path = rdata["default"]["longest_path"]["value"]
 
	print(f"plant n{index}")
	print(f"- branch_pts {branch_pts} points")
	print(f"- tips {tips} points")
	print(f"- area {area} px")
	print(f"- solidity {solidity} px")
	print(f"- perimeter {perimeter} px")
	print(f"- width {width} px")
	print(f"- height {height} px")
	print(f"- longest_path {longest_path} px\n")
```
